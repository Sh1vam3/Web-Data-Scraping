EXTRACTING TEXT FROM THE TAGS IN WEB PAGES USING PYTHON

```python
lines
```

```python
for l in lines:
    print(l.text)
```

```python
s=soup.find("div",class_ = "container")
```

```python
s
```

```python
lines_1=s.find_all("p")
```
